```python
import torchaudio
from transformers import Speech2TextProcessor, Speech2TextForConditionalGeneration
from IPython.display import Audio
```

```python
# Path to the uploaded .wav file
audio_file_path = '/content/harvard.wav'
```

```python
# Load the processor and model
processor = Speech2TextProcessor.from_pretrained("facebook/s2t-small-librispeech-asr")
model = Speech2TextForConditionalGeneration.from_pretrained("facebook/s2t-small-librispeech-asr")

# Load the original audio
input_audio, input_sampling_rate = torchaudio.load(audio_file_path, normalize=True)

# Define the desired sampling rate
desired_sampling_rate = 16000

# Resample to the desired sampling rate
resampler = torchaudio.transforms.Resample(input_sampling_rate, desired_sampling_rate, dtype=input_audio.dtype)
input_audio = resampler(input_audio)

# Prepare the input for the model
inputs = processor(input_audio.squeeze().numpy(), sampling_rate=desired_sampling_rate, return_tensors="pt")
```

```python
# Generate transcription
generated_ids = model.generate(inputs["input_features"])
transcription = processor.batch_decode(generated_ids, skip_special_tokens=True)
print("Transcription:", transcription[0])
```
